## 15. Distribution of price for each room type

```python
df['room_type'].unique()
```

```python
df.price = df.price.str[1:-3]
df.price = df.price.str.replace(",", "")
df.price = df.price.astype('float')
```

```python
room_price = df[['room_type','price']]
```

```python
entire = room_price[room_price['room_type']=='Entire home/apt']
```

```python
plt.figure(figsize=(20,4))
entire.price.hist(bins=100, range=(0,1000))
plt.margins(x=0)
plt.show()
```

```python
shared = room_price[room_price['room_type']=='Shared room']
```

```python
plt.figure(figsize=(20,4))
shared.price.hist(bins=100, range=(0,1000))
plt.margins(x=0)
plt.show()
```

```python
private = room_price[room_price['room_type']=='Private room']
plt.figure(figsize=(20,4))
private.price.hist(bins=100, range=(0,1000))
plt.margins(x=0)
plt.show()
```

```python
hotel = room_price[room_price['room_type']=='Hotel room']
plt.figure(figsize=(20,4))
hotel.price.hist(bins=100, range=(0,1000))
plt.margins(x=0)
plt.show()
```
